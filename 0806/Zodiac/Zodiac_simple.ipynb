```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%pylab inline

year = int(input("Input your birth year: "))

# Finish the full zodiac, hint: Dragon, Snake, Horse, Sheep, Rooster, Dog, Pig, Rat, Ox, Tiger
if (year - 2000) % 12 == 0:
    zodiac = 'Dragon'
elif (year - 2000) % 12 == 1:
    zodiac = 'Snake'
elif (year - 2000) % 12 == 2:
    zodiac = 'Horse'
else:
    zodiac = 'Hare'
    
print("Your Zodiac sign :",zodiac)

img=mpimg.imread("./files/" + zodiac +".jpeg")
imgplot = plt.imshow(img)
plt.show()
```
